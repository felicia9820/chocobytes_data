```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
%matplotlib inline

mob_df = pd.read_csv("mobility.csv")
bus_df = pd.read_csv("train.csv", sep=';', encoding='latin-1')
```

```python
mob_df
```

```python
bus_df
```

```python
bus_df['Provincia'] = bus_df['Provincia'].str.title()
bus_df
```

```python
bus_df[['Provincia', 'Regione']].value_counts()
```

```python
pbus_df = bus_df.groupby(['Provincia','Regione']).count().reset_index()
pbus_df
```

```python
merged = mob_df.merge(pbus_df, on=['Provincia'])
merged
```

```python
plt.scatter(merged['Comune'], merged['AUM'])
plt.xlabel('Number of bus stops')
plt.ylabel('Upward mobility')
m, b = np.polyfit(merged['Comune'], merged['AUM'], 1)
plt.plot(merged['Comune'], m*merged['Comune'] + b)
```

```python
rbus_df = pbus_df.groupby(by='Regione').sum().reset_index()
rbus_df.columns = ['Somma' if x=='Comune' else x for x in rbus_df.columns]
rbus_df
```

```python
rmob_df = merged.groupby(by='Regione').mean().reset_index()
rmob_df
```

```python
plt.scatter(rmob_df['Comune'], rmob_df['AUM'])
plt.xlabel('Number of bus stops')
plt.ylabel('Upward mobility')
m, b = np.polyfit(rmob_df['Comune'], rmob_df['AUM'], 1)
plt.plot(rmob_df['Comune'], m*rmob_df['Comune'] + b)
```

```python
merged_2 = rmob_df.merge(rbus_df, on=['Regione'])
merged_2
```

```python
plt.scatter(merged_2['Somma'], merged_2['AUM'])
plt.xlabel('Number of bus stops')
plt.ylabel('Upward mobility')
m, b = np.polyfit(merged_2['Somma'], merged_2['AUM'], 1)
plt.plot(merged_2['Somma'], m*merged_2['Somma'] + b)
```

```python
cbus_df = bus_df.groupby(['Comune','Provincia']).count().reset_index()
cbus_df
```

```python
cpbus_df = cbus_df.groupby(by='Provincia').mean().reset_index()
cpbus_df
```

```python
merged_3 = mob_df.merge(cpbus_df, on=['Provincia'])
merged_3
```

```python
plt.scatter(merged_3['Regione'], merged_3['AUM'])
plt.xlabel('Number of bus stops')
plt.ylabel('Upward mobility')
m, b = np.polyfit(merged_3['Regione'], merged_3['AUM'], 1)
plt.plot(merged_3['Regione'], m*merged_3['Regione'] + b)
```
